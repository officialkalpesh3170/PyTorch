```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

Encoding Y as it has all categorical data
**bold text**



```python
encoder = LabelEncoder()
y_train= encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_test
```

**CONVERTING NUMPY ARRAY TO TENSORS**

```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

*Defining the model*

```python
class MysimpleNN():
#initializing
  def __init__(self,X):
    self.weights=torch.rand(X.shape[1],1,dtype=torch.float64, requires_grad=True)
    self.bias =torch.zeros(1,dtype=torch.float64,requires_grad=True)
#Forward pass


  def forward(self,X):
    X = X.to(torch.float64)
    z=torch.matmul(X,self.weights)+self.bias
    return torch.sigmoid(z)
# Loss

  def loss_function(self,y_pred,y):
    epsilon=1e-7
    y_pred=torch.clamp(y_pred,epsilon,1-epsilon)
    loss=-(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred))
    return loss.mean()
```

```python
lr=0.1
epochs=300
```

```python
model1 =MysimpleNN(X_train_tensor)
```

```python
model1.weights
```

# Forward Pass

```python

for epoch in range(epochs):

  # forward pass
  y_pred = model1.forward(X_train_tensor)

  # loss calculate
  loss = model1.loss_function(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():
    model1.weights -=lr * model1.weights.grad
    model1.bias -= lr * model1.bias.grad

  # zero gradients
  model1.weights.grad.zero_()
  model1.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')

```

```python
model1.weights
```

# Y_prediction


```python
# model evaluation
with torch.no_grad():
  y_pred = model1.forward(X_test_tensor)
  y_pred = (y_pred > 0.47).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
